# Modeling Data

 1.  What is Data Loading in Power BI and why is it considered the first step of analysis?
- Data Loading in Power BI refers to the process of importing data from various sources (like Excel, databases, or web) into Power BI for analysis.

Why it is the first step:
Foundation of Analysis: Without data, no analysis can be performed.
Data Availability: Makes raw data accessible for transformation and visualization.
Integration: Combines data from multiple sources into one place.
Preparation for Next Steps: Enables cleaning, modeling, and reporting.

Hence, it is the starting point of the entire data analysis process.

2. Explain the difference between “Load” and “Transform Data” in Power BI.
- | Aspect      | Load                                | Transform Data                                |
| ----------- | ----------------------------------- | --------------------------------------------- |
| Meaning     | Directly imports data into Power BI | Opens Power Query Editor to clean/modify data |
| Usage       | When data is already clean          | When data needs preprocessing                 |
| Speed       | Faster                              | Takes more time                               |
| Flexibility | Limited                             | High flexibility                              |
| Tool Used   | Direct loading                      | Power Query Editor                            |

In simple terms:
Load = Import data as it is
Transform Data = Clean and shape data before loading

3. What is a Fact Table and a Dimension Table? Give examples from the dataset.
- Fact Table and Dimension Table
1. Fact Table
Contains quantitative (numerical) data
Used for analysis and calculations

Examples:

Sales Amount
Property Price
Number of Units Sold

In a Real Estate dataset:

Table with Price, Area, Sales Value → Fact Table
2. Dimension Table
Contains descriptive (qualitative) data
Provides context to fact data

Examples:

Owner Name
Location
Property Type

In a Real Estate dataset:

Table with Owner Name, City, Property Type → Dimension Table

4. Why is Star Schema preferred over Snowflake Schema in Power BI?
- Star Schema
Central Fact Table connected to Dimension Tables
Simple structure
Snowflake Schema
More complex with normalized dimension tables
Reasons Star Schema is preferred:
Better Performance
Faster query execution
Simpler Design
Easy to understand and use
Efficient Relationships
Fewer joins required
Optimized for Power BI
Works well with Power BI’s in-memory engine
Improved Visualization
Easier report creation
Conclusion:

Star Schema is preferred because it is simple, fast, and efficient for analysis in Power BI.

6. Identify and replace null values in Vaccination_Status
-  Steps in Power Query:
Open Power BI Desktop
Click on Home → Transform Data (opens Power Query Editor)
In the table, locate and select the column:
Vaccination_Status
Identify Null Values:
Scroll through the column and look for values labeled null
OR use filter dropdown → uncheck all → select only (null)

- Replace Null Values:
Method 1: Replace Values
Right-click on Vaccination_Status column
Select Replace Values
In the dialog box:
Value to Find → null
Replace With → "Not Vaccinated" (or any appropriate value like "Unknown")
Click OK

- Method 2: Using Transform Tab
Go to Transform → Replace Values
Enter:
Find → null
Replace → "Not Vaccinated"
Click OK

Final Step:
Click Close & Apply to save changes
